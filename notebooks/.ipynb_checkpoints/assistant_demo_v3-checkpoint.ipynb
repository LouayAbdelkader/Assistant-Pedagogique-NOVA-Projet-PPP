# 🔍 Moteur de Recherche Sémantique avec Catégorisation du Type de Question

## Objectif

Ce notebook implémente un assistant QA pédagogique qui, à partir d'une question posée par un étudiant :

1. **Catégorise le type de la question** (définition, formule, exemple, remarque, image)
2. **Filtre les slides** qui correspondent à ce type
3. **Extrait la meilleure réponse** dans ces slides
4. **Répond "je ne connais pas la réponse"** si aucun slide pertinent n'est trouvé

## Pourquoi catégoriser le type de question ?

Chaque slide du cours est étiqueté avec une catégorie :

| Catégorie | Exemples de questions |
|---|---|
| `definition` | "Qu'est-ce que… ?", "Définir…", "C'est quoi… ?" |
| `formule` | "Quelle est la formule de… ?", "Comment calcule-t-on… ?" |
| `exemple` | "Donne un exemple de…", "Illustrer… par un exemple" |
| `remarque` | "Quelles sont les propriétés de… ?", "Que remarque-t-on… ?" |
| `image` | "Montre le schéma de…", "Représentation graphique de…" |

En filtrant d'abord par type, on réduit le bruit et on améliore la précision de la réponse.

## Pipeline

```
Question étudiant
       │
       ▼
┌─────────────────────────┐
│ 1. Catégorisation       │  → type détecté : "definition" / "formule" / …
└────────────┬────────────┘
             │
             ▼
┌─────────────────────────┐
│ 2. Filtrage des slides  │  → garder uniquement les slides du bon type
└────────────┬────────────┘
             │
             ▼
┌─────────────────────────┐
│ 3. Recherche sémantique │  → similarité cosinus sur les slides filtrés
└────────────┬────────────┘
             │
             ▼
┌─────────────────────────┐
│ 4. Seuil de confiance   │  → score trop bas ? → "Je ne connais pas la réponse"
└────────────┬────────────┘
             │
             ▼
┌─────────────────────────┐
│ 5. Extraction de phrase │  → meilleure phrase dans le slide retenu
└────────────┬────────────┘
             │
             ▼
         Réponse finale
```

## Fichiers requis

| Fichier | Contenu |
|---|---|
| `data/3-numerical/slide_embeddings.npy` | Matrice (N_slides × 768) des embeddings pré-calculés |
| `data/3-numerical/slides_metadata.json` | Métadonnées : texte, chap_id, section, **category** |

## 1. Imports et Chargement du Modèle

````python
import json
import re
import os
from pathlib import Path
from collections import defaultdict

import numpy as np

# ── Désactivation des barres tqdm widgets (avant sentence_transformers) ───
os.environ["TQDM_DISABLE"] = "1"
import tqdm.auto
from tqdm import tqdm as tqdm_std
tqdm.auto.tqdm = tqdm_std

from sentence_transformers import SentenceTransformer

MODEL_NAME = "intfloat/multilingual-e5-base"
print(f"⏳ Chargement du modèle '{MODEL_NAME}'…")
model = SentenceTransformer(MODEL_NAME)
print(f"✅ Modèle chargé — {model.get_sentence_embedding_dimension()} dimensions")
````

## 2. Chargement des Données

Chaque slide dans `slides_metadata.json` doit contenir un champ `"category"` parmi :
`definition`, `formule`, `exemple`, `remarque`, `image`.

> ⚠️ Si ce champ est absent des métadonnées, la cellule suivante affiche un avertissement et assigne `"definition"` par défaut.

````python
EMBEDDINGS_PATH = Path("../data/3-numerical/slide_embeddings.npy")
METADATA_PATH   = Path("../data/3-numerical/slides_metadata.json")

for p in [EMBEDDINGS_PATH, METADATA_PATH]:
    if not p.exists():
        raise FileNotFoundError(f"Fichier manquant : {p}")

# ── Chargement ────────────────────────────────────────────────────────────
slide_embeddings = np.load(EMBEDDINGS_PATH)
N_slides, emb_dim = slide_embeddings.shape
print(f"📐 Embeddings : {N_slides} slides × {emb_dim} dimensions")

with open(METADATA_PATH, "r", encoding="utf-8") as f:
    slides_metadata = json.load(f)
print(f"📋 Métadonnées : {len(slides_metadata)} slides")

assert N_slides == len(slides_metadata), (
    f"Incohérence : {N_slides} embeddings mais {len(slides_metadata)} métadonnées"
)

# ── Vérification du champ 'category' ─────────────────────────────────────
CATEGORIES_VALIDES = {"definition", "formule", "exemple", "remarque", "image"}
sans_categorie = [i for i, s in enumerate(slides_metadata) if "category" not in s]

if sans_categorie:
    print(f"⚠️  {len(sans_categorie)} slide(s) sans champ 'category' → 'definition' assigné par défaut")
    for i in sans_categorie:
        slides_metadata[i]["category"] = "definition"
else:
    print("✅ Tous les slides ont un champ 'category'")

# ── Résumé des catégories présentes ──────────────────────────────────────
print("\nRépartition des slides par catégorie :")
comptage = defaultdict(int)
for s in slides_metadata:
    comptage[s["category"]] += 1
for cat, n in sorted(comptage.items()):
    print(f"  {cat:<12} : {n} slides")
````

## 3. Construction de l'Index par Catégorie

On pré-calcule pour chaque catégorie :
- la liste des **indices globaux** des slides qui lui appartiennent
- la **sous-matrice d'embeddings** correspondante

Cela évite de filtrer à la volée à chaque requête.

````python
# Regrouper les indices de slides par catégorie
cat_to_indices: dict[str, list[int]] = defaultdict(list)
for i, slide in enumerate(slides_metadata):
    cat_to_indices[slide["category"]].append(i)

# Pré-extraire les sous-matrices d'embeddings par catégorie
cat_index: dict[str, dict] = {}
for cat, indices in cat_to_indices.items():
    cat_index[cat] = {
        "indices"    : indices,
        "embeddings" : slide_embeddings[indices],  # sous-matrice (n × 768)
    }

print("🗂️  Index par catégorie :")
for cat, idx in cat_index.items():
    print(f"  {cat:<12} : {len(idx['indices'])} slides  "
          f"— sous-matrice {idx['embeddings'].shape}")
````

## 4. Catégorisation du Type de Question

### Principe

On définit pour chaque catégorie un **ensemble de phrases représentatives** (patrons linguistiques). On encode ces phrases et on calcule leur similarité avec la question de l'étudiant. La catégorie dont les phrases sont les plus proches sémantiquement est retenue.

```
Question : "Quelle est la formule de l'énergie ?"
   │
   ├── sim(Q, patrons_definition) = 0.61
   ├── sim(Q, patrons_formule)    = 0.89  ← max → catégorie = "formule"
   ├── sim(Q, patrons_exemple)    = 0.52
   ├── sim(Q, patrons_remarque)   = 0.55
   └── sim(Q, patrons_image)      = 0.48
```

Les embeddings des patrons sont calculés **une seule fois** au démarrage.

````python
# ── Patrons linguistiques par catégorie ───────────────────────────────────
# Chaque catégorie est représentée par plusieurs formulations typiques.
# Plus les patrons sont variés, meilleure est la détection.
CATEGORY_PATTERNS: dict[str, list[str]] = {
    "definition": [
        "Qu'est-ce que c'est ?",
        "Définir ce concept",
        "Donner la définition",
        "C'est quoi ?",
        "Qu'est-ce qu'on appelle ?",
        "Que signifie ce terme ?",
    ],
    "formule": [
        "Quelle est la formule mathématique ?",
        "Comment calcule-t-on ?",
        "Donner l'expression mathématique",
        "Quelle est l'équation ?",
        "Comment s'exprime mathématiquement ?",
        "Quelle est la relation entre ces grandeurs ?",
    ],
    "exemple": [
        "Donner un exemple concret",
        "Illustrer par un exemple",
        "Montrer un cas pratique",
        "Exemple d'application",
        "Citer un exemple",
    ],
    "remarque": [
        "Quelles sont les propriétés importantes ?",
        "Que remarque-t-on ?",
        "Quelles sont les caractéristiques ?",
        "Quelles sont les conditions ?",
        "Points importants à retenir",
        "Quelles sont les limites ou particularités ?",
    ],
    "image": [
        "Montrer le schéma",
        "Représentation graphique",
        "Tracer la courbe",
        "Illustration visuelle",
        "Afficher le diagramme",
        "Représenter sur un graphe",
    ],
}

# ── Pré-encodage des patrons (une seule fois) ─────────────────────────────
# Pour chaque catégorie, on encode tous ses patrons et on garde
# le vecteur MOYEN → représente le "centre sémantique" de la catégorie
category_embeddings: dict[str, np.ndarray] = {}
for cat, patterns in CATEGORY_PATTERNS.items():
    embs = model.encode(
        [f"query: {p}" for p in patterns],
        normalize_embeddings=True,
        show_progress_bar=False,
    )
    # Moyenne + re-normalisation → vecteur centroïde de la catégorie
    centroid = embs.mean(axis=0)
    centroid /= np.linalg.norm(centroid)  # re-normaliser après la moyenne
    category_embeddings[cat] = centroid

CATEGORY_NAMES = list(category_embeddings.keys())
# Matrice (N_categories × 768) pour le calcul vectorisé
category_matrix = np.stack([category_embeddings[c] for c in CATEGORY_NAMES])

print(f"✅ {len(CATEGORY_NAMES)} catégories encodées : {CATEGORY_NAMES}")


# ── Fonction de catégorisation ────────────────────────────────────────────
def categoriser_question(question: str, verbose: bool = False) -> dict:
    """
    Détermine le type d'une question (definition, formule, exemple, remarque, image).

    Compare l'embedding de la question au centroïde de chaque catégorie.
    Retourne la catégorie avec le score de similarité le plus élevé.

    Args:
        question:  Texte de la question de l'étudiant
        verbose:   Si True, affiche les scores de toutes les catégories

    Returns:
        {
          'categorie':  nom de la catégorie détectée (str),
          'score':      score de similarité avec cette catégorie (float),
          'all_scores': dict de tous les scores par catégorie
        }
    """
    q_emb = model.encode(
        [f"query: {question}"],
        normalize_embeddings=True,
        show_progress_bar=False,
    )
    # Similarité avec chaque centroïde de catégorie
    scores = np.dot(category_matrix, q_emb.T).flatten()

    best_i    = int(scores.argmax())
    all_scores = {CATEGORY_NAMES[i]: round(float(scores[i]), 4)
                  for i in range(len(CATEGORY_NAMES))}

    if verbose:
        print("Scores par catégorie :")
        for cat, sc in sorted(all_scores.items(), key=lambda x: -x[1]):
            bar = "█" * int(sc * 30)
            marker = " ← détectée" if cat == CATEGORY_NAMES[best_i] else ""
            print(f"  {cat:<12} {sc:.4f}  {bar}{marker}")

    return {
        "categorie" : CATEGORY_NAMES[best_i],
        "score"     : round(float(scores[best_i]), 4),
        "all_scores": all_scores,
    }


# ── Test rapide ───────────────────────────────────────────────────────────
print("\n🧪 Tests de catégorisation :\n")
questions_test_cat = [
    "Qu'est-ce qu'un signal périodique ?",
    "Quelle est la formule du produit de convolution ?",
    "Donne un exemple de signal à énergie finie",
    "Quelles sont les propriétés d'un système linéaire ?",
    "Montre le schéma d'un filtre passe-bas",
]
for q in questions_test_cat:
    cat = categoriser_question(q)
    print(f"  '{q}'")
    print(f"   → Catégorie : {cat['categorie']}  (score={cat['score']})\n")
````

## 5. Fonctions de Recherche

````python
def split_sentences(text: str) -> list[str]:
    """
    Découpe un texte en liste de phrases.

    Utilise un lookbehind sur la ponctuation (.!?) suivie d'un espace.
    """
    sentences = re.split(r'(?<=[.!?])\s+', text.strip())
    return [re.sub(r'\s+', ' ', s).strip() for s in sentences if s.strip()]
````

### `repondre` — Recherche Filtrée par Type de Question

Deux seuils de confiance contrôlent la qualité de la réponse :

| Seuil | Rôle |
|---|---|
| `seuil_categorie` | Score minimum pour faire confiance à la catégorie détectée. En dessous → recherche globale (tous les slides). |
| `seuil_reponse` | Score minimum de similarité entre la question et le meilleur slide. En dessous → "Je ne connais pas la réponse". |

````python
# Seuil en dessous duquel on répond "je ne sais pas"
SEUIL_REPONSE  = 0.60
# Seuil en dessous duquel la catégorie est jugée peu fiable → fallback global
SEUIL_CATEGORIE = 0.55

REPONSE_INCONNUE = "Je ne connais pas la réponse à cette question dans le contenu du cours."


def repondre(
    question: str,
    top_k: int = 1,
    use_sentence_extraction: bool = True,
    verbose: bool = False,
) -> dict:
    """
    Retourne la meilleure réponse pour une question, en filtrant par type.

    Pipeline :
      1. Catégoriser le type de la question
      2. Filtrer les slides par ce type (ou global si confiance faible)
      3. Chercher le slide le plus proche (similarité cosinus)
      4. Vérifier le seuil de confiance de la réponse
      5. Extraire la meilleure phrase dans le slide retenu

    Args:
        question:               Texte de la question
        top_k:                  Nombre de slides candidats à considérer
        use_sentence_extraction: Si True, extrait la meilleure phrase du slide
        verbose:                Affiche les étapes intermédiaires

    Returns:
        Dictionnaire avec :
          categorie, categorie_score, mode_recherche,
          chap_id, chap_name, slide_id, section,
          score, reponse, connue (bool)
    """

    # ── Étape 1 : Catégorisation du type de question ──────────────────────
    cat_result = categoriser_question(question, verbose=verbose)
    categorie       = cat_result["categorie"]
    categorie_score = cat_result["score"]

    if verbose:
        print(f"\n📂 Type détecté : '{categorie}'  (score={categorie_score})")

    # ── Étape 2 : Sélection de la zone de recherche ───────────────────────
    if categorie_score >= SEUIL_CATEGORIE and categorie in cat_index:
        # Recherche filtrée : seulement les slides du type détecté
        search_indices = cat_index[categorie]["indices"]
        search_matrix  = cat_index[categorie]["embeddings"]
        mode_recherche = f"filtré (type='{categorie}')"
    else:
        # Fallback : recherche sur tous les slides
        search_indices = list(range(N_slides))
        search_matrix  = slide_embeddings
        mode_recherche = "global (fallback — catégorie peu fiable)"

    if verbose:
        print(f"🔎 Mode : {mode_recherche}  →  {len(search_indices)} slides examinés")

    # ── Étape 3 : Encodage de la question ─────────────────────────────────
    q_emb = model.encode(
        [f"query: {question}"],
        normalize_embeddings=True,
        show_progress_bar=False,
    )

    # ── Étape 4 : Similarité cosinus sur la zone filtrée ──────────────────
    local_scores = np.dot(search_matrix, q_emb.T).flatten()
    local_best   = local_scores.argsort()[::-1][:top_k]

    best_local_idx  = local_best[0]
    best_score      = float(local_scores[best_local_idx])
    best_global_idx = search_indices[best_local_idx]

    if verbose:
        print(f"🎯 Meilleur score : {best_score:.4f}  "
              f"(seuil={SEUIL_REPONSE} → "
              f"{'réponse trouvée' if best_score >= SEUIL_REPONSE else 'SOUS LE SEUIL'})")

    # ── Étape 5 : Vérification du seuil de confiance ──────────────────────
    # Si le meilleur score est trop bas, la question n'a pas de réponse dans le cours
    if best_score < SEUIL_REPONSE:
        return {
            "categorie"      : categorie,
            "categorie_score": round(categorie_score, 4),
            "mode_recherche" : mode_recherche,
            "chap_id"        : None,
            "chap_name"      : None,
            "slide_id"       : None,
            "section"        : None,
            "score"          : round(best_score, 4),
            "reponse"        : REPONSE_INCONNUE,
            "connue"         : False,
        }

    # ── Étape 6 : Extraction de la réponse ───────────────────────────────
    slide   = slides_metadata[best_global_idx]
    content = slide.get("slide_content", "")

    if not content:
        reponse = REPONSE_INCONNUE
        connue  = False

    elif use_sentence_extraction:
        phrases = split_sentences(content)
        if not phrases:
            reponse = content[:300]
        else:
            phr_embs   = model.encode(
                [f"passage: {p}" for p in phrases],
                normalize_embeddings=True,
                show_progress_bar=False,
            )
            phr_scores = np.dot(phr_embs, q_emb.T).flatten()
            reponse    = phrases[int(phr_scores.argmax())]
        connue = True

    else:
        reponse = content[:500] + ("…" if len(content) > 500 else "")
        connue  = True

    return {
        "categorie"      : categorie,
        "categorie_score": round(categorie_score, 4),
        "mode_recherche" : mode_recherche,
        "chap_id"        : slide["chap_id"],
        "chap_name"      : slide["chap_name"],
        "slide_id"       : slide["slide_id"],
        "section"        : slide["section"],
        "score"          : round(best_score, 4),
        "reponse"        : reponse,
        "connue"         : connue,
    }
````

## 6. Tests et Démonstration

````python
def tester(question: str, verbose: bool = False) -> None:
    """Affiche la réponse formatée pour une question."""
    rep = repondre(question, verbose=verbose)
    print(f"\n{'─' * 62}")
    print(f"❓ Question          : {question}")
    print(f"🏷️  Type détecté      : {rep['categorie']}  "
          f"(confiance={rep['categorie_score']})")
    print(f"🔎 Mode de recherche : {rep['mode_recherche']}")

    if not rep["connue"]:
        # Réponse inconnue : affichage simplifié
        print(f"💬 Réponse           : ❌ {rep['reponse']}")
        print(f"   Score similarité  : {rep['score']}  (< seuil {SEUIL_REPONSE})")
    else:
        print(f"📚 Chapitre {rep['chap_id']}        : {rep['chap_name']}")
        print(f"📄 Slide {rep['slide_id']:>3}          : {rep['section']}")
        print(f"🎯 Score similarité  : {rep['score']}")
        print(f"💬 Réponse           : {rep['reponse']}")
````

````python
# ── Questions normales (réponses connues) ─────────────────────────────────
print("=" * 62)
print(" QUESTIONS AVEC RÉPONSES DANS LE COURS")
print("=" * 62)

tester("Qu'est-ce qu'un signal périodique ?")
tester("Quelle est la formule du produit de convolution ?")
tester("Définition de l'énergie d'un signal")
tester("C'est quoi mono-variable ?")
````

````python
# ── Questions hors cours (réponses inconnues) ─────────────────────────────
print("=" * 62)
print(" QUESTIONS HORS DU CONTENU DU COURS")
print("=" * 62)

tester("Quelle est la capitale de la France ?")        # hors cours
tester("Comment fonctionne un moteur à combustion ?")  # hors cours
tester("Qui a inventé le téléphone ?")                  # hors cours
````

## 7. Visualisation des Catégories Détectées

On teste un ensemble de questions variées et on affiche la distribution des types détectés.

````python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

questions_viz = [
    ("Qu'est-ce qu'un signal périodique ?",          "definition"),
    ("Définir produit de convolution",               "definition"),
    ("Définition de l'énergie d'un signal",          "definition"),
    ("C'est quoi mono-variable ?",                   "definition"),
    ("Quelle est la formule de l'énergie ?",         "formule"),
    ("Comment calcule-t-on la puissance ?",          "formule"),
    ("Donner un exemple de signal causal",           "exemple"),
    ("Quelles sont les propriétés d'un SLI ?",       "remarque"),
    ("Montrer le schéma d'un filtre",                "image"),
]

COLORS = {
    "definition": "#3498db",
    "formule"   : "#e74c3c",
    "exemple"   : "#2ecc71",
    "remarque"  : "#f39c12",
    "image"     : "#9b59b6",
}

resultats_viz = []
for q, attendu in questions_viz:
    cat = categoriser_question(q)
    resultats_viz.append({
        "question" : q[:45] + ("…" if len(q) > 45 else ""),
        "detecte"  : cat["categorie"],
        "score"    : cat["score"],
        "attendu"  : attendu,
        "correct"  : cat["categorie"] == attendu,
    })

# ── Graphique ─────────────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(11, 5))
y_pos = range(len(resultats_viz))

for i, r in enumerate(resultats_viz):
    color  = COLORS.get(r["detecte"], "#95a5a6")
    marker = "✅" if r["correct"] else "❌"
    ax.barh(i, r["score"], color=color, alpha=0.85, edgecolor="white")
    ax.text(r["score"] + 0.005, i,
            f"{marker} {r['detecte']} ({r['score']:.3f})",
            va="center", fontsize=8.5)

ax.set_yticks(list(y_pos))
ax.set_yticklabels([r["question"] for r in resultats_viz], fontsize=8)
ax.set_xlim(0, 1.15)
ax.axvline(SEUIL_CATEGORIE, color="gray", linestyle="--",
           linewidth=1.2, label=f"Seuil catégorie ({SEUIL_CATEGORIE})")
ax.set_xlabel("Score de similarité avec le type détecté")
ax.set_title("Catégorisation des questions par type", fontsize=12, fontweight="bold")

# Légende des couleurs
patches = [mpatches.Patch(color=c, label=cat) for cat, c in COLORS.items()]
ax.legend(handles=patches + [plt.Line2D([0], [0], color="gray",
          linestyle="--", label=f"Seuil ({SEUIL_CATEGORIE})")],
          loc="lower right", fontsize=8)

ax.spines[["top", "right"]].set_visible(False)
plt.tight_layout()
plt.show()

n_correct = sum(r["correct"] for r in resultats_viz)
print(f"\nPrécision catégorisation : {n_correct}/{len(resultats_viz)} "
      f"({100*n_correct//len(resultats_viz)}%)")
````

## 8. Récapitulatif

### Ce que fait ce notebook

1. **Catégorise** la question selon son type : `definition`, `formule`, `exemple`, `remarque`, `image`
2. **Filtre** les slides correspondant à ce type avant de chercher
3. **Répond "Je ne connais pas"** si le meilleur score est sous le seuil `SEUIL_REPONSE`
4. **Extrait** la phrase la plus pertinente dans le slide retenu

### Paramètres à ajuster

| Paramètre | Valeur par défaut | Effet |
|---|---|---|
| `SEUIL_REPONSE` | `0.60` | Seuil en dessous duquel on répond "je ne sais pas" — baisser = plus de réponses, risque de bruit |
| `SEUIL_CATEGORIE` | `0.55` | Seuil de confiance du type détecté — baisser = plus de fallback global |
| `CATEGORY_PATTERNS` | 5-6 patrons/catégorie | Ajouter des patrons améliore la détection |

### Pistes d'amélioration

- **Enrichir les patrons** : ajouter des formulations arabes ou anglaises pour les étudiants bilingues
- **Catégories imbriquées** : `definition_formelle` vs `definition_intuitive`
- **RAG génératif** : utiliser un LLM pour reformuler la phrase extraite en réponse naturelle
- **Journalisation** : enregistrer les questions sans réponse pour améliorer le cours
