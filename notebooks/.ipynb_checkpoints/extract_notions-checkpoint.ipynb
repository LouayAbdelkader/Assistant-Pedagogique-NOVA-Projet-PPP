```python
import re
#version améliorée
def extract_notions(chapter_name, section_title, slide_text, top_k=3, threshold=0.4):
    text_to_analyze = slide_text if len(slide_text) > 10 else section_title
    # spaCy : candidats linguistiquement valides
    doc = nlp(text_to_analyze)
    candidates = []
    for chunk in doc.noun_chunks:
        words = [token.lemma_ for token in chunk if not token.is_stop and not token.is_punct]
        text = " ".join(words).strip()
        text = chunk.text.replace('\n', ' ').strip()#enlève les espaces inutiles au début et à la fin
        text = re.sub(r'[^a-zA-ZÀ-ÿ\s]', '', text).strip()#gérer les sumboles, parenthèses, virgules

        if (len(text) > 3 and
            not nlp.vocab[text.lower()].is_stop and #on élimine les textes courts, les stop words
            not any(x in text.lower() for x in ['http', 'illustration', 'figure'])):    #on élimine les termes liés aux liens web
            candidates.append(text)

     # éviter les doublons
    candidates = sorted(list(set(candidates)), key=len, reverse=True)
    final_candidates = []
    for c in candidates:
      # On vérifie si le candidat est une répétition d'un concept déjà accepté
        is_redundant = False
        for accepted in final_candidates:
            # Si "signaux" est dans "signaux déterminstes", on considère que c'est redondant
            if c.lower() in accepted.lower() or accepted.lower() in c.lower():
                is_redundant = True
                break

        if not is_redundant:
            final_candidates.append(c)

    # on envoie uniquement 'final_candidates' au modèle E5
    if not final_candidates:
        # Si le contenu est vide, on renvoie les mots significatifs du titre directement
       return [w.text for w in nlp(section_title) if not w.is_stop and len(w.text) > 3][:top_k]


        # contexte multi-niveaux
    # (on limite le contenu pour éviter bruit + coût)
    short_content = " ".join(slide_text.split()[:50])
    # E5 formatting (IMPORTANT)
    query = f"query: {chapter_name}. {section_title}. {short_content}"
    passages = [f"passage: {c}" for c in final_candidates]

    # embeddings
    query_emb = model.encode([query], normalize_embeddings=True)
    passage_embs = model.encode(passages, normalize_embeddings=True)

    # similarité
    scores = np.dot(passage_embs, query_emb.T).flatten()

    # ranking : sort by relevance
    ranked = sorted(zip(candidates, scores), key=lambda x: x[1], reverse=True)

    # On augmente légèrement le threshold à 0.40 pour être plus sélectif
    notions = [text for text, score in ranked if score >= threshold][:top_k]

    return notions

```
