# 🔍 Moteur de Recherche Sémantique avec Catégorisation des Questions

## Contexte

Ce notebook implémente un assistant QA (Question Answering) qui, à partir d'une question, retrouve la réponse la plus pertinente dans un corpus de slides de cours.

## Nouveauté : Catégorisation avant Recherche

La recherche naïve compare la question à **tous les slides** du corpus. Avec la catégorisation, on réduit d'abord le sous-espace de recherche :

```
Question
   │
   ▼
┌─────────────────────────────┐
│  Étape 1 : Catégorisation   │  ← Quel chapitre / thème ?
└────────────┬────────────────┘
             │  ex: "Chapitre 2 – Systèmes linéaires"
             ▼
┌─────────────────────────────┐
│  Étape 2 : Recherche filtrée│  ← Comparaison sur ~15 slides au lieu de 75
└────────────┬────────────────┘
             │
             ▼
┌─────────────────────────────┐
│  Étape 3 : Extraction phrase│  ← Re-ranking au niveau phrase
└────────────┬────────────────┘
             │
             ▼
         Réponse finale
```

## Bénéfices

| Critère | Sans catégorisation | Avec catégorisation |
|---|---|---|
| Slides comparés | 75 (tous) | ~15 (1 chapitre) |
| Bruit de recherche | Élevé | Faible |
| Précision | Bonne | Meilleure |
| Scalabilité | Limitée | Extensible |

## Fichiers requis

| Fichier | Contenu |
|---|---|
| `data/3-numerical/slide_embeddings.npy` | Matrice (N_slides × 768) |
| `data/3-numerical/slides_metadata.json` | Métadonnées textuelles des slides |
| `data/2-json/cours.json` | Structure du cours (chapitres + titres) |

## 1. Imports et Chargement du Modèle

On utilise `multilingual-e5-base` : un bi-encodeur multilingue avec 768 dimensions.
Les préfixes `"query: "` et `"passage: "` sont **obligatoires** pour ce modèle.

````python
import json
import re
import os
import time
from pathlib import Path
from collections import defaultdict

import numpy as np

# ── Désactivation tqdm widgets (doit précéder sentence_transformers) ──────
os.environ["TQDM_DISABLE"] = "1"
import tqdm.auto
from tqdm import tqdm as tqdm_std
tqdm.auto.tqdm = tqdm_std

from sentence_transformers import SentenceTransformer

MODEL_NAME = "intfloat/multilingual-e5-base"
print(f"⏳ Chargement du modèle '{MODEL_NAME}'…")
model = SentenceTransformer(MODEL_NAME)
EMB_DIM = model.get_sentence_embedding_dimension()
print(f"✅ Modèle chargé — {EMB_DIM} dimensions")
````

## 2. Chargement des Données

On charge trois sources :
- **slide_embeddings.npy** : les vecteurs pré-calculés de chaque slide
- **slides_metadata.json** : les textes et métadonnées (chap_id, section…)
- **cours.json** : la structure du cours pour récupérer les **titres officiels des chapitres** utilisés lors de la catégorisation

````python
EMBEDDINGS_PATH = Path("../data/3-numerical/slide_embeddings.npy")
METADATA_PATH   = Path("../data/3-numerical/slides_metadata.json")
COURSE_PATH     = Path("../data/2-json/cours.json")

for p in [EMBEDDINGS_PATH, METADATA_PATH, COURSE_PATH]:
    if not p.exists():
        raise FileNotFoundError(f"Fichier manquant : {p}")

# ── Embeddings ────────────────────────────────────────────────────────────
slide_embeddings = np.load(EMBEDDINGS_PATH)
N_slides, emb_dim = slide_embeddings.shape
print(f"📐 Embeddings : {N_slides} slides × {emb_dim} dimensions")

# ── Métadonnées ───────────────────────────────────────────────────────────
with open(METADATA_PATH, "r", encoding="utf-8") as f:
    slides_metadata = json.load(f)
print(f"📋 Métadonnées : {len(slides_metadata)} slides")

# ── Structure du cours ────────────────────────────────────────────────────
with open(COURSE_PATH, "r", encoding="utf-8") as f:
    course = json.load(f)

# Dictionnaire chap_id → chap_name (ex: {1: "Représentation temporelle…", 2: "Systèmes…"})
chapitres = {chap["chap_id"]: chap["chap_name"] for chap in course["chapters"]}
print(f"📚 Cours : {len(chapitres)} chapitres")

assert N_slides == len(slides_metadata), "Incohérence embeddings / métadonnées"
print("✅ Cohérence vérifiée")

print("\nChapitres du cours :")
for cid, cname in chapitres.items():
    n = sum(1 for s in slides_metadata if s["chap_id"] == cid)
    print(f"  Chapitre {cid} : {cname}  ({n} slides)")
````

## 3. Construction de l'Index par Chapitre

Pour rendre la recherche filtrée efficace, on pré-calcule pour chaque chapitre :
- la liste des **indices** des slides qui lui appartiennent
- la **sous-matrice d'embeddings** correspondante (extraction de lignes)

Ainsi, lors d'une requête, on n'accède qu'à la sous-matrice du chapitre identifié — pas besoin de filtrer à la volée.

````python
# ── Index : chap_id → {indices, embeddings} ───────────────────────────────
# Regrouper les indices de chaque slide par chapitre
chap_to_indices: dict[int, list[int]] = defaultdict(list)
for i, slide in enumerate(slides_metadata):
    chap_to_indices[slide["chap_id"]].append(i)

# Pré-extraire les sous-matrices d'embeddings pour chaque chapitre
chap_index: dict[int, dict] = {}
for chap_id, indices in chap_to_indices.items():
    chap_index[chap_id] = {
        "indices"    : indices,
        "embeddings" : slide_embeddings[indices],   # sous-matrice (n_slides_du_chap × 768)
    }

print("🗂️  Index par chapitre :")
for cid, idx in chap_index.items():
    print(f"  Chapitre {cid} : {len(idx['indices'])} slides  "
          f"(sous-matrice {idx['embeddings'].shape})")
````

## 4. Catégorisation de la Question

### Principe

On encode les **titres officiels des chapitres** avec le préfixe `"passage: "`, puis on calcule leur similarité avec la question encodée en `"query: "`. Le chapitre dont le titre est le plus proche sémantiquement de la question est sélectionné.

```
Question : "Définir produit de convolution"
   │
   ▼  encode("query: Définir produit de convolution")  → vecteur Q
   │
   ├── cos_sim(Q, encode("passage: Représentation temporelle…")) = 0.61
   ├── cos_sim(Q, encode("passage: Systèmes linéaires continus")) = 0.84  ← max
   └── cos_sim(Q, encode("passage: Transformée de Fourier"))      = 0.72
                                              → Chapitre 2 sélectionné
```

### Optimisation : encodage des titres une seule fois

Les embeddings des titres de chapitres sont calculés **une fois au démarrage** et réutilisés pour toutes les requêtes.

````python
# ── Pré-encodage des titres de chapitres ──────────────────────────────────
# Calculé une seule fois — réutilisé pour toutes les requêtes
chap_ids    = list(chapitres.keys())
chap_names  = list(chapitres.values())

chap_title_embeddings = model.encode(
    [f"passage: {name}" for name in chap_names],
    normalize_embeddings=True,
    show_progress_bar=False,
)  # shape : (N_chapitres, 768)

print(f"✅ {len(chap_ids)} titres de chapitres encodés")
for cid, cname in zip(chap_ids, chap_names):
    print(f"  [{cid}] {cname}")


# ── Fonction de catégorisation ────────────────────────────────────────────
def categoriser(question: str, seuil_confiance: float = 0.0) -> dict:
    """
    Identifie le chapitre le plus pertinent pour une question.

    Compare l'embedding de la question à ceux des titres de chapitres.
    Retourne le chapitre avec le score de similarité le plus élevé.

    Args:
        question:          Texte de la question
        seuil_confiance:   Score minimum requis (0 = pas de seuil)

    Returns:
        {
          'chap_id':    identifiant du chapitre (int),
          'chap_name':  nom du chapitre (str),
          'score':      score de similarité (float, entre 0 et 1),
          'confiant':   True si le score dépasse le seuil (bool)
        }
    """
    # Encoder la question
    q_emb = model.encode(
        [f"query: {question}"],
        normalize_embeddings=True,
        show_progress_bar=False,
    )

    # Similarité avec chaque titre de chapitre
    scores = np.dot(chap_title_embeddings, q_emb.T).flatten()

    best_i  = int(scores.argmax())
    best_id = chap_ids[best_i]

    return {
        "chap_id"  : best_id,
        "chap_name": chap_names[best_i],
        "score"    : round(float(scores[best_i]), 4),
        "confiant" : float(scores[best_i]) >= seuil_confiance,
        "all_scores": {chap_ids[i]: round(float(scores[i]), 4) for i in range(len(chap_ids))},
    }


# ── Test rapide ───────────────────────────────────────────────────────────
print("\n🧪 Test de catégorisation :")
for q in ["Définir produit de convolution", "Qu'est-ce qu'un signal périodique ?"]:
    cat = categoriser(q)
    print(f"  '{q}'")
    print(f"   → Chapitre {cat['chap_id']} : {cat['chap_name']}  (score={cat['score']})")
    print(f"     Scores par chapitre : {cat['all_scores']}")
````

## 5. Fonctions de Recherche

### `split_sentences` — Segmentation en phrases

````python
def split_sentences(text: str) -> list[str]:
    """
    Découpe un texte en liste de phrases.

    Utilise un lookbehind sur la ponctuation (.!?) suivi d'un espace.
    Évite de couper sur des abréviations (fig. 3, e.g., etc.).

    Args:
        text: texte brut
    Returns:
        Liste de phrases nettoyées
    """
    sentences = re.split(r'(?<=[.!?])\s+', text.strip())
    return [re.sub(r'\s+', ' ', s).strip() for s in sentences if s.strip()]
````

### `repondre_avec_categorie` — Recherche Filtrée par Chapitre

Cette fonction combine les deux étapes :

1. **Catégorisation** → identifie le chapitre cible via `categoriser()`
2. **Recherche filtrée** → compare la question uniquement aux slides de ce chapitre
3. **Extraction de phrase** → re-ranking au niveau phrase dans le meilleur slide

Un paramètre `fallback_global` permet de revenir à la recherche globale si le score de catégorisation est trop faible (question hors-chapitres connus).

````python
def repondre_avec_categorie(
    question: str,
    top_k: int = 1,
    use_sentence_extraction: bool = True,
    seuil_categorie: float = 0.55,
    verbose: bool = False,
) -> dict:
    """
    Recherche sémantique accélérée avec catégorisation préalable.

    Pipeline :
        1. Catégoriser la question → chapitre cible
        2. Chercher uniquement dans les slides de ce chapitre
        3. Extraire la meilleure phrase (si use_sentence_extraction=True)
        4. Fallback global si score de catégorisation < seuil_categorie

    Args:
        question:               Texte de la question
        top_k:                  Nombre de slides candidats (dans le chapitre filtré)
        use_sentence_extraction: Si True, extrait la meilleure phrase du slide
        seuil_categorie:        Score minimum pour faire confiance à la catégorisation ;
                                en dessous → fallback recherche globale
        verbose:                Affiche les détails intermédiaires

    Returns:
        Dictionnaire avec : chap_id, chap_name, slide_id, section,
                            score, reponse, categorie_score, mode_recherche
    """

    # ── Étape 1 : Catégorisation ──────────────────────────────────────────
    cat = categoriser(question)
    categorie_score = cat["score"]

    if verbose:
        print(f"📂 Catégorie détectée : Chapitre {cat['chap_id']} – {cat['chap_name']}")
        print(f"   Score catégorisation : {categorie_score}  "
              f"({'✅ confiant' if categorie_score >= seuil_categorie else '⚠️ faible → fallback global'})")
        print(f"   Scores : {cat['all_scores']}")

    # ── Étape 2 : Sélection de la zone de recherche ───────────────────────
    if categorie_score >= seuil_categorie:
        # Recherche filtrée : seulement les slides du chapitre identifié
        target_chap_id  = cat["chap_id"]
        search_indices  = chap_index[target_chap_id]["indices"]
        search_matrix   = chap_index[target_chap_id]["embeddings"]
        mode_recherche  = f"filtré (chapitre {target_chap_id})"
    else:
        # Fallback : recherche sur tous les slides
        search_indices  = list(range(N_slides))
        search_matrix   = slide_embeddings
        mode_recherche  = "global (fallback)"

    if verbose:
        print(f"   Mode : {mode_recherche}  →  {len(search_indices)} slides examinés")

    # ── Étape 3 : Encodage de la question ─────────────────────────────────
    q_emb = model.encode(
        [f"query: {question}"],
        normalize_embeddings=True,
        show_progress_bar=False,
    )

    # ── Étape 4 : Similarité cosinus sur la zone filtrée ──────────────────
    # scores[i] = similarité entre la question et search_indices[i]
    local_scores = np.dot(search_matrix, q_emb.T).flatten()

    # Indices locaux des top_k meilleurs (dans la zone filtrée)
    local_best = local_scores.argsort()[::-1][:top_k]

    if verbose:
        print(f"   Top-{top_k} scores locaux : {local_scores[local_best].round(4)}")

    resultats = []
    for local_idx in local_best:
        # Reconvertir l'indice local en indice global
        global_idx = search_indices[local_idx]
        slide      = slides_metadata[global_idx]
        content    = slide.get("slide_content", "")
        score      = float(local_scores[local_idx])

        # ── Étape 5 : Extraction de la réponse ───────────────────────────
        if not content:
            reponse = "[Aucun texte exploitable dans ce slide]"

        elif use_sentence_extraction:
            phrases = split_sentences(content)
            if not phrases:
                reponse = content[:300]
            else:
                phr_embs = model.encode(
                    [f"passage: {p}" for p in phrases],
                    normalize_embeddings=True,
                    show_progress_bar=False,
                )
                phr_scores = np.dot(phr_embs, q_emb.T).flatten()
                reponse = phrases[int(phr_scores.argmax())]
        else:
            reponse = content[:500] + ("…" if len(content) > 500 else "")

        resultats.append({
            "chap_id"         : slide["chap_id"],
            "chap_name"       : slide["chap_name"],
            "slide_id"        : slide["slide_id"],
            "section"         : slide["section"],
            "score"           : round(score, 4),
            "reponse"         : reponse,
            "categorie_score" : round(categorie_score, 4),
            "mode_recherche"  : mode_recherche,
        })

    return resultats[0] if resultats else {}
````

## 6. Tests et Démonstration

### Affichage formaté

````python
def tester(question: str, verbose: bool = False) -> None:
    """Affiche la réponse formatée pour une question."""
    rep = repondre_avec_categorie(question, verbose=verbose)
    if not rep:
        print(f"⚠️  Aucune réponse trouvée pour : '{question}'")
        return
    print(f"\n{'─' * 60}")
    print(f"❓ Question         : {question}")
    print(f"📂 Catégorie        : Chap. {rep['chap_id']} – {rep['chap_name']}  "
          f"(score cat.={rep['categorie_score']})")
    print(f"🔎 Mode recherche   : {rep['mode_recherche']}")
    print(f"📄 Slide {rep['slide_id']:>3}          : {rep['section']}")
    print(f"🎯 Score similarité : {rep['score']}")
    print(f"💬 Réponse          : {rep['reponse']}")
````

````python
if __name__ == "__main__":
    questions = [
        "Qu'est-ce qu'un signal périodique ?",
        "Définir produit de convolution",
        "Définition de l'énergie d'un signal",
        "C'est quoi mono-variable ?",
    ]
    for q in questions:
        tester(q)
````

### Mode verbose — Voir les détails internes

````python
# Exemple avec verbose=True pour voir toutes les étapes intermédiaires
tester("Définir produit de convolution", verbose=True)
````

## 7. Benchmark : Recherche Filtrée vs Globale

On compare les deux approches sur le même jeu de questions pour mesurer :
- Le **gain de temps** (moins de slides à comparer)
- La **précision** (la réponse change-t-elle ?)

````python
import matplotlib.pyplot as plt

# ── Version globale (ancienne) ────────────────────────────────────────────
def repondre_global(question: str, use_sentence_extraction: bool = True) -> dict:
    """Version sans catégorisation — recherche sur tous les slides."""
    q_emb  = model.encode([f"query: {question}"], normalize_embeddings=True)
    scores = np.dot(slide_embeddings, q_emb.T).flatten()
    idx    = int(scores.argmax())
    slide  = slides_metadata[idx]
    content = slide.get("slide_content", "")
    if not content:
        reponse = "[Aucun texte]"
    elif use_sentence_extraction:
        phrases = split_sentences(content)
        if not phrases:
            reponse = content[:300]
        else:
            phr_embs   = model.encode([f"passage: {p}" for p in phrases], normalize_embeddings=True)
            phr_scores = np.dot(phr_embs, q_emb.T).flatten()
            reponse = phrases[int(phr_scores.argmax())]
    else:
        reponse = content[:500]
    return {"slide_id": slide["slide_id"], "score": round(float(scores[idx]), 4), "reponse": reponse}


# ── Benchmark ─────────────────────────────────────────────────────────────
questions_bench = [
    "Qu'est-ce qu'un signal périodique ?",
    "Définir produit de convolution",
    "Définition de l'énergie d'un signal",
    "C'est quoi mono-variable ?",
]

N_REPEAT = 5   # répétitions pour moyenner les temps
resultats_bench = []

for q in questions_bench:
    # Temps méthode globale
    t0 = time.perf_counter()
    for _ in range(N_REPEAT): r_glob = repondre_global(q)
    t_glob = (time.perf_counter() - t0) / N_REPEAT

    # Temps méthode filtrée
    t0 = time.perf_counter()
    for _ in range(N_REPEAT): r_filt = repondre_avec_categorie(q)
    t_filt = (time.perf_counter() - t0) / N_REPEAT

    resultats_bench.append({
        "question"   : q[:40] + "…",
        "t_global"   : round(t_glob * 1000, 1),    # ms
        "t_filtre"   : round(t_filt * 1000, 1),    # ms
        "gain_pct"   : round((t_glob - t_filt) / t_glob * 100, 1),
        "meme_rep"   : r_glob["reponse"] == r_filt["reponse"],
    })

import pandas as pd
df_bench = pd.DataFrame(resultats_bench)
print(df_bench.to_string(index=False))

# ── Graphique comparatif ──────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(9, 4))
x = range(len(questions_bench))
w = 0.35
b1 = ax.bar([i - w/2 for i in x], df_bench["t_global"], width=w,
            label="Recherche globale",  color="#e74c3c", alpha=0.85)
b2 = ax.bar([i + w/2 for i in x], df_bench["t_filtre"],  width=w,
            label="Recherche filtrée", color="#2ecc71", alpha=0.85)

# Annotations gain
for i, row in df_bench.iterrows():
    ax.text(i, max(row["t_global"], row["t_filtre"]) + 1,
            f"-{row['gain_pct']}%", ha="center", fontsize=9,
            color="#27ae60", fontweight="bold")

ax.set_xticks(list(x))
ax.set_xticklabels([q[:30] + "…" for q in questions_bench], rotation=15, ha="right", fontsize=8)
ax.set_ylabel("Temps (ms)")
ax.set_title("Benchmark : recherche globale vs filtrée par catégorie",
             fontsize=12, fontweight="bold")
ax.legend()
ax.spines[["top", "right"]].set_visible(False)
plt.tight_layout()
plt.show()
````

## 8. Récapitulatif

### Pipeline complet

```
Question
  │
  ├─ categoriser()     →  Chapitre cible  +  score de confiance
  │                        │
  │                        ├─ score ≥ seuil  →  sous-matrice du chapitre
  │                        └─ score < seuil  →  matrice globale (fallback)
  │
  ├─ dot(q_emb, matrix)  →  scores cosinus
  ├─ top-k slides         →  meilleur(s) slide(s)
  └─ split + re-rank      →  meilleure phrase  →  Réponse
```

### Ce qui a été ajouté par rapport à la version initiale

| Nouveauté | Bénéfice |
|---|---|
| `categoriser()` | Identifie le chapitre avant de chercher |
| `chap_index` pré-calculé | Accès O(1) aux sous-matrices par chapitre |
| `seuil_categorie` + fallback | Robustesse aux questions hors-corpus |
| `mode_recherche` dans le résultat | Traçabilité (filtré vs global) |
| `categorie_score` dans le résultat | Indicateur de confiance exposé |
| Benchmark temporel | Preuve du gain de performance |

### Pistes d'amélioration

- **Enrichir les titres** : ajouter des mots-clés ou une description au titre de chapitre pour améliorer la catégorisation
- **Catégorisation multi-label** : retourner les 2 chapitres les plus proches si les scores sont proches
- **RAG génératif** : brancher un LLM (ex. Mistral, LLaMA) pour reformuler la réponse extraite
- **Cache des embeddings** de questions fréquentes pour éliminer le temps d'encodage
