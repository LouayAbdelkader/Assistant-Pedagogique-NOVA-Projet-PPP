# Objectif :

À partir des logs d'interaction entre les utilisateurs et l'assistant, on calcule des indicateurs permettant à l'enseignant de suivre l'engagement et la continuité d'apprentissage.

Les indicateurs sont :
- Nombre de questions par chapitre
- Répétition des mêmes notions ou questions

> **Note** : en l'absence temporaire d'une vraie base de logs, on génère un jeu de données synthétique réaliste pour illustrer les calculs.

# Génération de logs synthétiques

```python
import json, os
from pprint import pprint

with open("../data/2-json/cours.json", "r", encoding="utf-8") as f:
    course = json.load(f)

# Agréger les slides de TOUS les chapitres (introduction ignorée : chap_id > 0)
slides = []
for chap in course["chapters"]:
    for slide in chap["content"]:
        slide["chap_id"]   = chap["chap_id"]
        slide["chap_name"] = chap["chap_name"]
        slides.append(slide)

print(f"{len(course['chapters'])} chapitre(s) chargé(s).")
print(f"{len(slides)} slides au total.")
```

```python
import random
from datetime import datetime, timedelta
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

random.seed(42)
np.random.seed(42)

# ── Paramètres de simulation ──────────────────────────────────────────────
STUDENTS      = [f"etudiant_{i:02d}" for i in range(1, 11)]   # 10 étudiants
# liste de dicts {section, chap_name, chap_id} pour la simulation
SLIDE_INFO    = [{"section": s["section"], "chap_name": s["chap_name"], "chap_id": s["chap_id"]} for s in slides]
SAMPLE_QUESTIONS = [
    "Qu'est-ce qu'un signal périodique ?",
    "Définition de l'énergie d'un signal",
    "Différence signal analogique et numérique ?",
    "Qu'est-ce que le spectre d'amplitude ?",
    "Qu'est-ce que la convolution ?",
    "Définition de la puissance d'un signal",
]

START_DATE = datetime(2025, 1, 15)

logs = []
for student in STUDENTS:
    n_sessions = random.randint(4, 12)
    for s in range(n_sessions):
        session_start = START_DATE + timedelta(
            days=random.randint(0, 60),
            hours=random.randint(8, 22)
        )
        session_duration_min = random.gauss(25, 10)   # durée en minutes
        session_duration_min = max(5, session_duration_min)
        n_questions = random.randint(1, 6)

        for q_idx in range(n_questions):
            # certains étudiants posent les mêmes questions (répétition)
            question = random.choices(
                SAMPLE_QUESTIONS,
                weights=[3,2,2,1,1,1,1,1],   # questions fréquentes plus probables
                k=1
            )[0]
            slide_info = random.choice(SLIDE_INFO)
            ts         = session_start + timedelta(minutes=q_idx * session_duration_min / n_questions)

            logs.append({
                "user_id"             : student,
                "session_id"          : f"{student}_s{s:02d}",
                "timestamp"           : ts.isoformat(),
                "section"             : slide_info["section"],
                "chap_name"           : slide_info["chap_name"],
                "chap_id"             : slide_info["chap_id"],
                "question"            : question,
                "session_duration_min": round(session_duration_min, 1)
            })

import pandas as pd
df_logs = pd.DataFrame(logs)
df_logs["timestamp"] = pd.to_datetime(df_logs["timestamp"], format="ISO8601")
df_logs["date"]      = df_logs["timestamp"].dt.date

print(f"{len(df_logs)} interactions générées pour {df_logs['user_id'].nunique()} étudiants.")
print(df_logs.head())
```

# Implémentation des indicateurs d'engagement

## Nombre de questions par section

Cet indicateur révèle **quelles sections du cours sont les plus consultées**, et donc potentiellement les plus difficiles pour les étudiants.

```python
questions_par_chap = (
    df_logs.groupby("section")["question"]
    .count()
    .sort_values(ascending=False)
    .rename("nb_questions")
    .reset_index()
)

#print("Nombre de questions par section :")
#print(questions_par_chap.to_string(index=False))

plt.figure(figsize=(10, 4))
plt.barh(questions_par_chap["section"][::-1],
         questions_par_chap["nb_questions"][::-1], color="coral")
plt.title("Nombre de questions posées par section du cours")
plt.xlabel("Nombre de questions")
plt.tight_layout()
plt.show()
```

## Répétition des mêmes notions ou questions

Si un étudiant pose plusieurs fois la même question (ou des variantes proches), cela indique que **la notion n'a pas été comprise**.

On détecte deux niveaux de répétition :
- **Répétition exacte** : même question posée plusieurs fois par le même étudiant.
- **Répétition sémantique** : questions proches détectées par similarité cosinus des embeddings.

```python
# ── Répétition exacte ────────────────────────────────────────────────────
repetitions_exactes = (
    df_logs.groupby(["user_id", "question"])["session_id"]
    .count()
    .rename("nb_fois")
    .reset_index()
)
repetitions_exactes = repetitions_exactes[repetitions_exactes["nb_fois"] > 1].sort_values(
    "nb_fois", ascending=False
)

print("Questions posées plusieurs fois par le même étudiant :")
print(repetitions_exactes.to_string(index=False))
```

```python
# Chargement du modèle
import os
os.environ["TQDM_DISABLE"] = "1"

# Force tqdm to use plain text mode - must happen before sentence_transformers import
from tqdm.auto import tqdm as tqdm_auto
from tqdm import tqdm as tqdm_std
import tqdm.auto
tqdm.auto.tqdm = tqdm_std  # replace notebook/widget tqdm with plain text tqdm

from sentence_transformers import SentenceTransformer

model = SentenceTransformer("intfloat/multilingual-e5-base")
```

```python
# ── Répétition sémantique (clustering par embeddings) ────────────────────

unique_questions = df_logs["question"].unique().tolist()
q_embeddings = model.encode(
    [f"query: {q}" for q in unique_questions],
    normalize_embeddings=True,
    show_progress_bar=False
)

# Matrice de similarité
sim_matrix = np.dot(q_embeddings, q_embeddings.T)

# Groupes de questions similaires (seuil > 0.85)
SIM_THRESHOLD = 0.85
groups = []
visited = set()

for i, q in enumerate(unique_questions):
    if i in visited:
        continue
    group = [q]
    visited.add(i)
    for j, q2 in enumerate(unique_questions):
        if j != i and j not in visited and sim_matrix[i, j] >= SIM_THRESHOLD:
            group.append(q2)
            visited.add(j)
    groups.append(group)

print("Groupes de questions sémantiquement proches (répétition conceptuelle) :")
for k, g in enumerate(groups):
    if len(g) > 1:
        print(f"  Groupe {k+1} : {g}")
    else:
        print(f"  Question isolée : {g[0]}")
```

# Synthèse des indicateurs d'engagement

On agrège tous les indicateurs dans un **tableau de bord par étudiant** utilisable par l'enseignant.


```python
# ── Tableau de bord synthétique par étudiant ─────────────────────────────

# Nb total de questions
nb_q = df_logs.groupby("user_id")["question"].count().rename("nb_questions_total")

# Nb de questions répétées
rep = (
    repetitions_exactes.groupby("user_id")["nb_fois"].sum().rename("nb_repetitions")
)

# Section la plus consultée
fav_section = (
    df_logs.groupby(["user_id", "chap_name"])["question"].count()
    .reset_index()
    .sort_values("question", ascending=False)
    .groupby("user_id").first()["chap_name"]
    .rename("chapitre_fav")
)

dashboard = pd.concat([nb_q, rep, fav_section], axis=1).fillna(0)
dashboard["nb_repetitions"] = dashboard["nb_repetitions"].astype(int)

print("=== Tableau de bord enseignant ===")
print(dashboard.to_string())

# Sauvegarde
dashboard.to_csv("../data/dashboard_engagement.csv")
print("\nSauvegardé dans data/dashboard_engagement.csv")
```
